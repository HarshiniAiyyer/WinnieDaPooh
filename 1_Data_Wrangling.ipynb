### Reading all the three datasets

Coz we like to read between the lines - I mean datasets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
dfams = pd.read_csv('https://raw.githubusercontent.com/HarshiniAiyyer/WinnieDaPooh/refs/heads/main/data/2021-05-18_honeypot-cloud-digitalocean-geo-1_netflow-v5.csv')
dfams.head()
```

```python
dffrank = pd.read_csv('https://raw.githubusercontent.com/HarshiniAiyyer/WinnieDaPooh/refs/heads/main/data/2021-05-18_honeypot-cloud-digitalocean-geo-3_netflow-v5.csv')
dffrank.head()
```

```python
dflon = pd.read_csv('https://raw.githubusercontent.com/HarshiniAiyyer/WinnieDaPooh/refs/heads/main/data/2021-05-18_honeypot-cloud-digitalocean-geo-4_netflow-v5.csv')
dflon.head()
```

### Adding location and concatenating the three of em together

One makes a stag, two makes a couple and three makes a dataset.

```python
dfams['Location'] = 'Amsterdam'
dffrank['Location'] = 'Frankfurt'
dflon['Location'] = 'London'
```

```python
df1 = pd.concat([dfams, dffrank, dflon])
```

### Dealing with null values

"Oh NaN NaN, what's my Name?"  ~ Not Rihanna

```python
# Calculate percentage of null values for each column
nulls = (df1.isnull().mean() * 100).round(2)

# Convert to dictionary for better readability
nullsdct = nulls.to_dict()

# Display the results
for col, percent in nullsdct.items():
    print(f"{col}: {percent}%")
```

### Dropping null values

There is a beauty in all correctness.

```python
df1 = df1.dropna().reset_index(drop=True)
```

```python
df1.info()
```

### Type Casting

Because I dont want a port address to be a floating point.

```python
import pandas as pd

# Convert to datetime
df1['StartTime'] = pd.to_datetime(df1['StartTime'])

# Convert to int (only if all values are numeric)
df1['Sport'] = pd.to_numeric(df1['Sport'], errors='coerce').fillna(-1).astype('int64')
df1['Dport'] = pd.to_numeric(df1['Dport'], errors='coerce').fillna(-1).astype('int64')
df1['dTos'] = pd.to_numeric(df1['dTos'], errors='coerce').fillna(-1).astype('int64')
df1['sTos'] = pd.to_numeric(df1['sTos'], errors='coerce').fillna(-1).astype('int64')


# Convert categorical fields
cat_cols = ['Proto', 'Dir', 'State', 'SrcAddr', 'DstAddr', 'Location']
df1[cat_cols] = df1[cat_cols].astype('category')
```

### Converting Hex Values found in port numbers to integers

Hex, not vex.

```python
def convert_to_int(val):
    try:
        if isinstance(val, str) and val.startswith('0x'):
            return int(val, 16)  # Convert hex to int
        else:
            return int(val)  # Already an int or numeric string
    except:
        return -1  # Sentinel for malformed values

# Apply conversion
df1['Sport'] = df1['Sport'].apply(convert_to_int)
df1['Dport'] = df1['Dport'].apply(convert_to_int)
```

### Sort Dataset based on the Earliest Date

Early Bear catches the Honeypot.

```python
df1 = df1.sort_values(by='StartTime').reset_index(drop=True)
df1.head()
```
